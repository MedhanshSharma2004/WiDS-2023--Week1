In this part of the assignment you have to implement multivariate gradient descent to find the minimas (local and global) of the given function:
Note : you can find different minimas by changing your initialisation.  
$f(x) = x^4 + x^2y^2 - y^2 + y^4 + 6$

```python
import numpy as np

# Function definition
def f(x, y):
    return x**4 + x**2*y**2 + y**4 - 6 - y**2

# Multivariate Gradient Descent
def gradient_descent(x_init, y_init, learning_rate, num_iterations):
    x = x_init
    y = y_init

    for i in range(num_iterations):
        # Calculate gradients
        grad_x = 4*x**3 + 2*x*y**2
        grad_y = 2*x**2*y + 4*y**3 - 2*y

        # Update variables using the gradient descent update rule
        x = x - learning_rate * grad_x
        y = y - learning_rate * grad_y

        # Print the current values and objective function value
        print(f"Iteration {i + 1}: x = {x}, y = {y}, f(x, y) = {f(x, y)}")

    return x, y, f(x, y)

# Initial values and hyperparameters
initial_x = 1.0
initial_y = 1.0
learning_rate = 0.01
num_iterations = 1200

# Run gradient descent
final_x, final_y, minima_value = gradient_descent(initial_x, initial_y, learning_rate, num_iterations)

print(f"\nFinal Minima: x = {final_x}, y = {final_y}, f(x, y) = {minima_value}")
```
